```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('default')

h = 6.626e-34 # Planckâ€™s constant
c = 299792458 # Speed of light
k = 1.380649e-23 # Boltzmann constant
```

```python
data = list(np.ones(200, dtype=np.int64) * 645) + list(np.ones(21, dtype=np.int64) * 510) + list(np.ones(232, dtype=np.int64) * 440)

plt.figure(figsize=(7,4))
plt.hist(data)
plt.xticks(np.arange(400, 701, 50), fontsize=15)
plt.xlabel("Wavelength [nm]", fontsize=20)
plt.yticks([])
plt.ylabel("Intensity", fontsize=20)
plt.savefig("purple_as_rgb")
plt.show()
```

```python
def blackbody(l, T, e, offset):
    return (e * ((2 * h * c**2) / l**5) * (1 / (np.exp((h * c) / (l * k * T)) - 1))) + offset

x = np.arange(400, 1001, 2)
y = blackbody(x * 1e-9, 5500, 1, 0) + np.random.normal(0, 2e11, len(x))

plt.figure(figsize=(7,4))
plt.scatter(x, y, s=2, color="blue")
plt.xticks(np.arange(400, 1001, 100), fontsize=15)
plt.xlabel("Wavelength [nm]", fontsize=20)
plt.yticks([])
plt.ylabel("Intensity", fontsize=20)
plt.show()
```

```python
def blackbody(l, T, e, offset):
    return (e * ((2 * h * c**2) / l**5) * (1 / (np.exp((h * c) / (l * k * T)) - 1))) + offset

x = np.arange(400, 1001, 2)
yt = blackbody(x * 1e-9, 5500, 1, 0)

plt.figure(figsize=(7,4))
plt.scatter(x, y, s=2, color="blue")
plt.plot(x, yt, linewidth=3, color="red")
plt.xticks(np.arange(400, 1001, 100), fontsize=15)
plt.xlabel("Wavelength [nm]", fontsize=20)
plt.yticks([])
plt.ylabel("Intensity", fontsize=20)
plt.show()
```
