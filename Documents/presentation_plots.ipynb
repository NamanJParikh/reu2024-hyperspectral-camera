```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3 * x + 90

def data(x):
    return f(x) + np.random.normal(0, 10)

xs = np.arange(0, 15, 1)
ys = np.array([data(x) for x in xs])
ys_exact = f(xs)

plt.figure(figsize=(5,5))
plt.scatter(xs, ys)
plt.plot(xs, ys_exact, c="red", linewidth=3)

plt.plot([xs[10], xs[10]], [ys[10], ys_exact[10]], c="green", linewidth=3)
plt.plot([xs[4], xs[4]], [ys[4], ys_exact[4]], c="green", linewidth=3)
plt.plot([xs[7], xs[7]], [ys[7], ys_exact[7]], c="green", linewidth=3)

plt.title("Linear Least Squares Regression", fontsize=18)
plt.show()
```
